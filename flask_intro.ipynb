```python
from flask import Flask , jsonify
import requests

app = Flask(__name__)

@app.route("/api/news",methods=['GET'])
def get_news():
  
  resp=requests.get(url)
  if resp.status_code == 200:
    news_data = resp.json()
    total_art=len(news_data['articles'])  #* to get the total number of articles
    first_art_author=news_data['articles'][0] #* to get the author of the first article
    first_art_title=news_data['articles'][0]['title']  # ** This will give you the title of the first article
    first_art_desc=news_data['articles'][0]['description']  # ** This will give you the description of the first article
    first_art_content=news_data['articles'][0]['content']  # ** This will give you the content of the first article 

    output_data = {
    "Total_articles": total_art,
    "Author": first_art_author,
    "Title": first_art_title,
    "Description": first_art_desc,
    "Content": first_art_content 
    }
    return jsonify(output_data)
  else:
    output_data = {
        "error": "Failed to fetch news data",
        "status_code": resp.status_code
    }
    return jsonify(output_data)
     
  return jsonify(output_data)

if __name__ == "__main__":
  app.run(debug=True, port=5000)  # ** Run the Flask app on port 5000

```
